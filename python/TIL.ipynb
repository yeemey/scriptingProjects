Finally learnt what it means to [stack](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.stack.html#pandas.DataFrame.stack) and [unstack](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) pandas data frames. I had missing data I needed to deal with, and [someone on StackOverflow had the same question](https://stackoverflow.com/questions/35181761/filling-in-missing-data-in-pandas-dataframe). Although they chose to go a different route, I went with the original suggestion since I'd been meaning to learn stack/unstack anyway. 

```python
import pandas as pd
import numpy as np
#toy data - the original was from csv
d = {'id': ['a', 'b', 'c', 'a', 'c', 'a', 'b', 'b', 'c', 'a', 'c', 'a', 'b'], 
        'sampling time': [100, 100, 100, 350, 350, 820, 820, 100, 100, 350, 350, 820, 820], 
        'event': ['evt1', 'evt1', 'evt1', 'evt1', 'evt1', 'evt1', 'evt1', 
                  'evt2', 'evt2', 'evt2', 'evt2', 'evt2', 'evt2'], 
        'frequency': np.random.rand(13)}
df = pd.DataFrame(d)
df
```

```python
pvt_df = df.pivot_table(index=['id', 'sampling time'], columns='event', values='frequency')
pvt_df
```

```python
# unstack() converts the indexes to columns, and imputes(?) missing values
pvt_df.unstack(level='sampling time')
```

```python
# after filling in missing values with 0, stack() "stacks" the columns back as indexes.
pvt_df.unstack(level='sampling time', fill_value=0).stack(level='sampling time')
```

```python
# if fill_value=None, which is the default, stack() will just get you the same df back, as if unstack() never happened.
pvt_df.unstack(level='sampling time').stack(level='sampling time')
```
